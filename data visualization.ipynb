```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
target = iris.target
```

```python
x_reduced = decomposition.PCA(n_components=3).fit_transform(iris.data)
```

```python
fig = plt.figure()
axes = Axes3D(fig)
axes.set_title('iris 3D')
axes.set_xlabel('x')
axes.set_ylabel('Y')
axes.set_zlabel('z')
axes.w_xaxis.set_ticklabels(())
axes.w_yaxis.set_ticklabels(())
axes.w_zaxis.set_ticklabels(())
axes.scatter(x_reduced[:,0],x_reduced[:,1],x_reduced[:,2],c=target)
```
